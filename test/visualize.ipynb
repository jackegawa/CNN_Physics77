# Visulize Loss & Accuracy

```python
import json
import matplotlib.pyplot as plt
```

## 1. Load Data

```python
# 1. Load PyTorch Benchmark
with open('pytorch_history.json', 'r') as f:
    torch_data = json.load(f)

# 2. Load Custom Model Data
# my_data = ...
```

## 2. Plotting

### 2.1. Accuracy

```python
ig, ax = plt.subplots(figsize=(8, 6))

ax.plot(torch_data['epoch'], torch_data['test_acc'], label='PyTorch (Benchmark)')
# ax.plot(my_data['epoch'], my_data['test_acc'], label='My Implementation')
ax.set_xlabel('Epoch')
ax.set_ylabel('Accuracy')
ax.legend()
ax.set_title('Test Accuracy Comparison')

plt.tight_layout()
plt.savefig("PyTorch_Accuracy.png")
plt.show()
```

### 2.2. Loss

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(torch_data['epoch'], torch_data['test_loss'], label='PyTorch (Benchmark)')
# ax.plot(my_data['epoch'], my_data['test_loss'], label='My Implementation')
ax.set_xlabel('Epoch')
ax.set_ylabel('Test Loss')
ax.legend()
ax.set_title('Test Loss Comparison')

plt.tight_layout()
plt.savefig("PyTorch_Loss.png")
plt.show()
```
